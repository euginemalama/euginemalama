```python
import numpy as np
import numpy.linalg as lg
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([(1,2),(1,3),(1,4),(1,5),(1,6),(2,3),(3,4),(4,5),(5,6),(6,2)])
plt.clf()
nx.draw(G)
plt.savefig("G.png")
plt.show()
```

```python
plt.clf()
pos=nx.spring_layout(G)
nx.draw_networkx_labels(G,pos)
nx.draw_networkx_nodes(G,pos)
nx.draw_networkx_edges(G,pos)
plt.axis('off')
plt.savefig("G.png")
plt.show()
```

```python
A=nx.to_numpy_matrix(G)
print(A)
```

```python
G2=nx.from_numpy_matrix(A)
plt.clf()
pos2=nx.spring_layout(G2)
nx.draw_networkx(G2,pos2)
plt.axis('off')
plt.savefig("G2.png")
plt.show()
```

```python
n = len(A)
e = np.ones(n)
k = np.dot(A,e)
print(k)
```

```python
H=G
H.remove_edges_from([(1,3),(4,5),(5,6)])
plt.clf()
nx.draw_networkx_labels(H,pos)
nx.draw_networkx_nodes(H,pos)
nx.draw_networkx_edges(H,pos)
plt.axis('off')
#nx.draw(H)
plt.savefig("H.png")
plt.show()
```

```python
C1 = [1,2,3]
colour=[]
for i in range(n):
    if i in C1:
        colour.append('r')
    else:
        colour.append('b')
        
plt.clf()
nx.draw_networkx(G2,pos2,node_color = colour)
plt.axis('off')
plt.savefig('ass.png')
plt.show()
```
