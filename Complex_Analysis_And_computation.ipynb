```python
import numpy as np
import numpy.linalg as lg
import networkx as nx
import matplotlib.pyplot as plt
```

```python
K=nx.read_weighted_edgelist('Neurons C elegans.txt')
```

```python
plt.clf()
nx.draw_networkx(K)
plt.axis('off')
plt.savefig("ass.png")
plt.show()
A_K=nx.to_numpy_matrix(K)
```

```python
print(A_K)
```

```python
#Redraw network graph to order node labelling in adjacency matrix.
K2=nx.from_numpy_matrix(A_K)
plt.clf()
pos=nx.spring_layout(K2)
nx.draw_networkx(K2)
#nx.draw_networkx_edges(K2,pos)
plt.axis('off')
plt.savefig("pth2.png")
plt.show()
#Define some vector quantities
n = len(A_K)
e = np.ones(n)
k = np.dot(A_K,e)
print(k)
```

```python
u1,v = lg.eig(A_K)
u1,v = lg.eigh(A_K)
print(u1)
```

```python
print(sum(u1**3))
print(sum(u1**4))
print(sum(k))
```

```python
156.0/34
```

```python
#nx.draw_networkx(K2)
nx.draw_networkx_edges(K2,pos)
nx.draw_networkx_nodes(K2,pos,node_size=10)
plt.axis('off')
plt.savefig("pth2.png")
plt.show()
```
